```python
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as pltpy

f = open("/workspace/VS/DataSet/Carbon dioxide (CO2) Emissions.csv", "r")
line = f.readline()
stati = []
while line != "":
    elments = line.split(",")
    if elments[0] not in stati:
        stati.append(elments[0])
    line = f.readline()
f.close()
#print(stati)


f = open("/workspace/VS/DataSet/Life expectancy at birth.csv", "r")
line = f.readline()
nations = []
y1990 = []
y2000 = []
y2012 = []
while line != "":
    elments = line.split(",")
    if elments[0] in stati and elments[2] == "Both sexes":    
        if elments[0] not in nations:
            nations.append(elments[0])
        match elments[1]:
            case "1990":
                y1990.append(int(elments[3]))
            case "2000":
                y2000.append(int(elments[3]))
            case "2012":
                y2012.append(int(elments[3]))
    line = f.readline()
f.close()

plt.rcParams["figure.figsize"] = (10, 10)
df = pd.DataFrame({'1990': y1990[:19], '2000': y2000[:19], '2012': y2012[:19]}, index = nations[:19])
ax = df.plot.barh()
pltpy.title("Life expectancy at birth")
pltpy.xlabel("years")
pltpy.xlim(60, 90)

df = pd.DataFrame({'1990': y1990[19:], '2000': y2000[19:], '2012': y2012[19:]}, index = nations[19:])
ax = df.plot.barh()
pltpy.title("Life expectancy at birth")
pltpy.xlabel("years")
pltpy.xlim(60, 90)
```
