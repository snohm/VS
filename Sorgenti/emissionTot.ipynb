```python
import matplotlib.pyplot as plt

nations = ["Germany", "Japan", "Russian Federation", "United States of America"]
years = []
for i in range (1990, 2021, 1): 
    years.append(i)
fCo2 = open("/workspace/VS/DataSet/Carbon dioxide in kilotonne.csv", "r")
fGh = open("/workspace/VS/DataSet/Greenhouse Gas in kilotonne CO2 equivalent.csv", "r")
tot = 0
val = []
label = ""
color = ["#1b9e77", "#d95f02", "#7570b3", "#e7298a"]
lineCo2 = fCo2.readline()
elmCo2 = lineCo2.split(",")
lineGh = fGh.readline()
elmGh = lineGh.split(",")
while lineCo2 != "":
    if elmCo2[0] in nations:
        for i in range (0, 31, 1):
            tot = tot + int(float(elmCo2[2])/1000) + int(float(elmGh[2])/1000)
            lineCo2 = fCo2.readline()
            elmCo2 = lineCo2.split(",")
            lineGh = fGh.readline()
            elmGh = lineGh.split(",")
        val.append(tot)
        tot = 0
    else:
        lineCo2 = fCo2.readline()
        elmCo2 = lineCo2.split(",")
        lineGh = fGh.readline()
        elmGh = lineGh.split(",")
fCo2.close()
fGh.close()
plt.rcParams["figure.figsize"] = (5, 5)
plt.pie(val, labels = nations, colors = color)
plt.title("Total emissions")
plt.show() 
```
