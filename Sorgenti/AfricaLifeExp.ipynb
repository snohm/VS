```python
import matplotlib.pyplot as plt

f = open("/workspace/VS/DataSet/StatesOfAfrica", "r")
line = f.readline()
africa = []
while line != "":
    africa.append(line[:len(line)-1])
    line = f.readline()
f.close()


f = open("/workspace/VS/DataSet/LifeExpGrosso.csv", "r")
years = [1980, 1988, 1990, 1998, 2000, 2010, 2021]
val = [0, 0, 0, 0, 0, 0, 0]
i = 6
label = ""
line = f.readline()
elments = line.split(",")
graphColor = "#f6f2ef"
plt.figure(facecolor = graphColor)
plt.axes(facecolor = graphColor)
while line != "":
    if elments[0] in africa:
        for j in range (0, 152, 1):
            if elments[1] == "1980" or elments[1] == "1988" or elments[1] == "1990" or elments[1] == "1998" or elments[1] == "2000" or elments[1] == "2010" or elments[1] == "2021":
                val[i] = int(float(elments[3]))
                i = i - 1
            label = elments[0]
            line = f.readline()
            elments = line.split(",")
        if val[0] > val[6]:
            plt.plot(years, val, label = label, color = 'blue')
        elif val[6] < 55:
            plt.plot(years, val, label = label, color = '#f03b20')
        elif val[6] > 75:    
           plt.plot(years, val, label = label, color = '#78c679')
        else: 
            plt.plot(years, val, color = 'gray')    
        val = [0, 0, 0, 0, 0, 0, 0] 
        i = 6
    else:
        line = f.readline()
        elments = line.split(",")
f.close()
plt.legend(ncol = 1, bbox_to_anchor=(1.02, 1), loc='upper left', facecolor = graphColor)
plt.title("Life expectancy: Africa")
plt.xlabel("years", loc='right')
plt.ylabel("age", rotation='horizontal', loc = 'top')
plt.ylim(0, 90)
plt.xticks(ticks= [1980,1990,2000,2010,2021], labels = ['1980','1990','2000','2010','2021'])
plt.show()
```
