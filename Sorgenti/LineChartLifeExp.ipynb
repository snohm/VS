```python
import matplotlib.pyplot as plt

f = open("/workspace/VS/DataSet/Life expectancy at birth.csv", "r")
years = ["1990", "2000", "2012"]
val = [0, 0, 0]
i = 2
label = ""
line = f.readline()
elments = line.split(",")
while line != "":
    for j in range (0, 9, 1):
        if elments[2] == "Both sexes":
            val[i] = int(elments[3])
            i = i - 1
        label = elments[0]
        line = f.readline()
        elments = line.split(",")
    plt.plot(years, val, label = label)
    val = [0, 0, 0] 
    i = 2
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left')
plt.show()    
f.close()
```
