```python
import matplotlib.pyplot as plt

f = open("/workspace/VS/DataSet/Carbon dioxide in kilotonne.csv", "r")
line = f.readline()
stati = []
while line != "":
    elments = line.split(",")
    if elments[0] not in stati:
        stati.append(elments[0])
    line = f.readline()
f.close()

f = open("/workspace/VS/DataSet/Life expectancy at birth.csv", "r")
line = f.readline()
nations = []
while line != "":
    elments = line.split(",")
    if elments[0] in stati and elments[0] not in nations:    
            nations.append(elments[0])
    line = f.readline()
f.close()


years = []
for i in range (2020, 1989, -1): 
    years.append(i)
f = open("/workspace/VS/DataSet/Hydrofluorocarbons in kilotonne CO2 equivalent.csv", "r")
val = []
label = ""
color = ["#d95f02", "#7570b3", "#e7298a"]
j = 0
line = f.readline()
elments = line.split(",")
while line != "" :
    #input()
    if elments[0] in nations:
       # print(line)
        #print(elments[0])
        prev = elments[0]
        add = True
        i = 0
        while prev == elments[0]:
            #if int(float(elments[2])/1000) < 0:
             #  line = f.readline()
             #  elments = line.split(",")
             #  add = False 
             #  break  
            val.append(int(float(elments[2])/1000)) # Mton
            i = i+1
            label = elments[0]
            prev = elments[0]
            line = f.readline()
            elments = line.split(",")
        #print(line[:len(line)-1])
        #print(label+" "+str(i))
        if add == True and val[0] > 20:
            plt.plot(years[:i], val, label = label, color = color[j])
            j = j+1
        elif add == True:
            plt.plot(years[:i], val, color = "gray")
        val = []
    else:
        line = f.readline()
        elments = line.split(",")    
f.close()
plt.title("Hydrofluorocarbons emissions")
plt.ylabel("MTon", rotation='horizontal', loc = 'top')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left')
plt.show() 
```
