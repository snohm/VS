```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import pandas as pd

f = open("/workspace/VS/DataSet/Undernourishment.csv", "r")
lifeExp = []
cases = []
line = f.readline()
elements = line.split(";")
graphColor = "#f6f2ef"
plt.figure(facecolor = graphColor)
plt.axes(facecolor = graphColor)

while line != "":
    elements = line.strip().split(";")
    if elements[0] == "Central African Republic" and elements[3] == "2001":
        for j in range(0, 20, 1):
            lifeExp.append(float(elements[1]))
            cases.append(float(elements[2]))
            line = f.readline()
            elements = line.strip().split(";")
    else:
        line = f.readline()
        elements = line.strip().split(";")

f.close()

years = []
for i in range (2001,2021,1):
    years.append(i)
    
df = pd.DataFrame({'y1': lifeExp, 'y2': cases, 'years': years })

plt.title("Undernourishment in\nCentral African Republic")
plt.xlabel("\nyears", loc='right')
plt.ylim(0, 60)
plt.yticks([0, 10, 20, 30, 40, 50, 60], labels = ['', '10', '20', '30', '40', '50', '60'])
plt.ylabel("age", rotation='horizontal', loc = 'top')
df['y1'].plot(kind='line')

df['y2'].plot(kind='bar', color='orange', secondary_y=True)
plt.ylim(0, 80)

# etichetta asse y2
plt.text(28, 82, "population percentage", rotation='horizontal', ha='right', va='center')
plt.yticks([0, 20, 40, 60, 80], labels = ['', '20%', '40%', '60%', '80%'])


# asse x
plt.xticks(ticks=[0, 6, 12, 19], labels =['', '', '', ''])
plt.text(0, -4, "2001", rotation='horizontal', ha='center')
plt.text(6,-4, "2007", rotation='horizontal', ha='center')
plt.text(12,-4, "2013", rotation='horizontal', ha='center')
plt.text(19,-4, "2021", rotation='horizontal', ha='center')

LE = mpatches.Patch(color='blue', label='Life Expectancy')
PP = mpatches.Patch(color='orange', label='Percentage of affected population')
plt.legend(handles=[LE, PP],ncol = 1, bbox_to_anchor=(1.12, 0.95), loc='upper left', facecolor = graphColor)

plt.show()
```
