```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import pandas as pd

f = open("/workspace/VS/DataSet/GHI.csv", "r")
lifeExp = []
ghi = []
line = f.readline()
elments = line.split(";")
graphColor = "#f6f2ef"
plt.figure(facecolor = graphColor)
plt.axes(facecolor = graphColor)
while line != "":
    if elments[0] == "Central African Republic":
        for j in range (0, 4, 1):
            lifeExp.append(int(float(elments[2])))
            ghi.append(int(float(elments[3])))
            line = f.readline()
            elments = line.split(";")
    else:
        line = f.readline()
        elments = line.split(";")
f.close()
years = [2000, 2008, 2015, 2021]


df = pd.DataFrame({'y1': lifeExp, 'y2': ghi, 'years': years })
plt.title("GHI in\nCentral African Republic")
plt.xlabel("\nyears", loc='right')
plt.ylim(0, 60)
plt.yticks([0,10, 20, 30, 40, 50, 60], labels = ['', '10', '20', '30', '40', '50', '60'])

plt.ylabel("age", rotation='horizontal', loc = 'top')
df['y1'].plot(kind='line')

df['y2'].plot(kind='line', color='orange', secondary_y=True)
plt.ylim(0, 60)

# etichetta asse y2
plt.text(3.6, 61, "score", rotation='horizontal', ha='right', va='center')

plt.yticks([0,10, 20, 30, 40, 50, 60], labels = ['', '10', '20', '30', '40', '50', '60'])

plt.xticks(ticks= [0, 1, 2, 3], labels = ['2000', '2008', '2015', '2021'])

LE = mpatches.Patch(color='blue', label='Life Expectancy')
PP = mpatches.Patch(color='orange', label='GHI')
plt.legend(handles=[LE, PP],ncol = 1, bbox_to_anchor=(1.5, 1), loc='upper right', facecolor = graphColor)

plt.show()
```
