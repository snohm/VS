```python
import matplotlib.pyplot as plt
import pandas as pd
from Diseases import CoutnCasesPercentage
from GiraVettori import rotate

f = open("/workspace/VS/DataSet/LifeExpGrosso.csv", "r")
lifeExp = []
line = f.readline()
elments = line.split(",")
graphColor = "#f0f0f0"
plt.figure(facecolor = graphColor)
plt.axes(facecolor = graphColor)
while line != "":
    if elments[0] == "Central African Republic" and elments[1] == "2021":
        for j in range (0, 42, 1):
            lifeExp.append(int(float(elments[3])))
            line = f.readline()
            elments = line.split(",")
    else:
        line = f.readline()
        elments = line.split(",")
f.close()
years = []
for i in range (1980,2022,1):
    years.append(i)
cases = CoutnCasesPercentage()
cases = cases[0:22]
for i in range (0, 20, 1):
    cases.append(0)
cases = rotate(cases)
lifeExp = rotate(lifeExp)


df = pd.DataFrame({'y1': lifeExp, 'y2': cases, 'years': years })
#fig, ax1 = plt.subplots(figsize=(15, 10))
#plt.rcParams["figure.facecolor"] = (graphColor)
#df.set_index('years', inplace= True)
plt.title("Diseases cases in\nCentral African Republic")
plt.xlabel("\nyears", loc='right')
plt.ylim(0, 90)
#plt.yticks([0,15,30,45,60], labels = ['','15','30','45','60'])
plt.ylabel("age", rotation='horizontal', loc = 'top')
df['y1'].plot(kind='line')
df['y2'].plot(kind='bar', color='y', secondary_y=True)
plt.ylim(20, 100)
#plt.ylabel("population percentage", rotation='horizontal', loc = 'top')
plt.yticks([20, 30, 40, 50, 60, 70, 80, 90, 100], labels = ['20%', '30%', '40%', '50%', '60%', '70%', '80%', '90%', '100%'])
#plt.xticks(ticks= [0,5,10,15,20,25,30,35,41], labels = ['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2015', '2021'])
plt.xticks(ticks= [0,5,10,15,20,25,30,35,41], labels = ['', '', '', '', '', '', '', '', ''])
#plt.xticks(rotation = 'horizontal')
plt.show()

```
