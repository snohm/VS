```python
import matplotlib.pyplot as plt
import pandas as pd
from Diseases import CoutnCasesPercentage
from GiraVettori import rotate

f = open("/workspace/VS/DataSet/LifeExpGrosso.csv", "r")
lifeExp = []
line = f.readline()
elments = line.split(",")
graphColor = "#f0f0f0"
while line != "":
    if elments[0] == "Central African Republic" and elments[1] == "2021":
        for j in range (0, 22, 1):
            lifeExp.append(int(float(elments[3])))
            line = f.readline()
            elments = line.split(",")
    else:
        line = f.readline()
        elments = line.split(",")
f.close()
years = []
for i in range (1980,2022,1):
    years.append(i)
lifeExp = rotate(lifeExp)
cases = CoutnCasesPercentage()
cases = rotate(cases[0:22])


df = pd.DataFrame({'y1': lifeExp, 'y2': cases })
fig, ax1 = plt.subplots(figsize=(15, 10))
plt.rcParams["figure.facecolor"] = (graphColor)
plt.rcParams["figure.edgecolor"] = (graphColor)
plt.ylim(0,100)
df['y1'].plot(kind='line', marker='d')
df['y2'].plot(kind='bar', color='y', secondary_y=True)

```
