```python
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as pltpy

f = open("/workspace/VS/DataSet/Carbon dioxide in kilotonne.csv", "r")
line = f.readline()
stati = []
while line != "":
    elments = line.split(",")
    if elments[0] not in stati:
        stati.append(elments[0])
    line = f.readline()
f.close()
#print(stati)

stati = ["Germany", "Japan", "Russian Federation", "United States of America"]

f = open("/workspace/VS/DataSet/Healthy life expectancy at birth.csv", "r")
line = f.readline()
nations = []
y2000 = []
y2012 = []
while line != "":
    elments = line.split(",")
    if elments[0] in stati and elments[2] == "Both sexes":    
        if elments[0] not in nations:
            nations.append(elments[0])
        match elments[1]:
            case "2000":
                y2000.append(int(elments[3]))
            case "2012":
                y2012.append(int(elments[3]))
    line = f.readline()
f.close()

plt.rcParams["figure.figsize"] = (7, 7)
df = pd.DataFrame({ '2000': y2000, '2012': y2012}, index = nations)
ax = df.plot.barh(color={"2000": "#b8e186", "2012":  "#d8b365"})
pltpy.title("Healy life expectancy at birth")
pltpy.xlabel("years")
pltpy.xlim(50, 80)
```
