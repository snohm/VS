```python
import matplotlib.pyplot as plt

f = open("/workspace/VS/DataSet/Carbon dioxide in kilotonne.csv", "r")
line = f.readline()
stati = []
while line != "":
    elments = line.split(",")
    if elments[0] not in stati:
        stati.append(elments[0])
    line = f.readline()
f.close()

f = open("/workspace/VS/DataSet/Life expectancy at birth.csv", "r")
line = f.readline()
nations = []
while line != "":
    elments = line.split(",")
    if elments[0] in stati and elments[0] not in nations:    
            nations.append(elments[0])
    line = f.readline()
f.close()


years = []
for i in range (1990, 2021, 1): 
    years.append(i)
f = open("/workspace/VS/DataSet/Carbon dioxide in kilotonne.csv", "r")
val = []
label = ""
color = ["#1b9e77", "#d95f02", "#7570b3", "#e7298a"]
j = 0
line = f.readline()
elments = line.split(",")
while line != "" :
    #prev = elments[0] 
    #input()
    if elments[0] in nations:
       # print(line)
        #print(elments[0])
        for i in range (0, 31, 1):
            #if int(float(elments[2])/1000) > 1000:
             #  line = f.readline()
             #  elments = line.split(",")
             #  break   
            val.append(int(float(elments[2])/1000)) # Mton
            #print(elments[1])
            label = elments[0]
            line = f.readline()
            elments = line.split(",")
        #print(len(years))
        #print(len(val))
        if i == 30 and val[len(val)-1] > 1000:
            plt.plot(years, val, label = label, color = color[j])
            j = j+1
        elif i == 30:
            plt.plot(years, val, color = "gray")
        val = []
        #i=i+1
    else:
        line = f.readline()
        elments = line.split(",")    
f.close()
#print(i)
plt.title("Carbon dioxide emissions")
plt.ylabel("MTon", rotation='horizontal', loc = 'top')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left')
plt.ylim(0, 8000)
plt.show() 
```
